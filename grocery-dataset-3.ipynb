```python

import keras_ocr
import matplotlib.pyplot as plt

```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
# List all files in the directory
directory = "/kaggle/input/grocery/freiburg_groceries_dataset/images/BEANS"
files = os.listdir(directory)

# Construct full paths to the images
file_paths = [os.path.join(directory, file) for file in files]

# Read the images using keras_ocr.tools.read()
Keras_images = [keras_ocr.tools.read(file_path) for file_path in file_paths[:10]]
```

```python
prediction_groups = pipeline.recognize(Keras_images)
```

```python
# Plot the predictions
fig, axs = plt.subplots(nrows=len(Keras_images), figsize=(20, 20))
for ax, image, predictions in zip(axs, Keras_images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```

```python
import os
from paddleocr import PaddleOCR

# Setup OCR model with English language
ocr_model = PaddleOCR(lang='en')

# List all files in the directory
directory = "/kaggle/input/d/saitejasri/beansimg"
files = os.listdir(directory)

# Construct full paths to the images
file_paths = [os.path.join(directory, file) for file in files]

for image_path in file_paths:  # Process only the first two images
    result = ocr_model.ocr(image_path)
    text = ""
    if result is not None:  # Check if result is not None
        for line in result:
            for word in line:
                text += word[1][0] + ' '  # Extract the text from OCR result
    else:
        print("No text detected in image:", image_path)
    print(text.strip())  # Print the extracted text (or empty string if no text detected)
```
